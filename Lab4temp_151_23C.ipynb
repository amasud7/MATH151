### Question 1

#### 1a

```python
#tangent line
x = symbols('x')
f = 8/((x**2)+4)
df = diff(f, x)
m = df.subs(x, 2)
tangent = m*(x-2) + f.subs(x, 2)
print(f"Equation of tangent line: {tangent}")
```

#### 1b

```python
# normal slope
m2 = -1/m
normal = m2*(x-2) + f.subs(x, 2)
print(f"Equation of normal line: {normal}")
```

#### 1c

```python
p1 = plot(f, (x, 0, 4), ylim = [0, 4], show=False) # graph of f
p2 = plot(tangent, (x, 0, 4), ylim = [0, 4], show=False) # graph of tangent 
p3 = plot(normal, (x, 0, 4), ylim = [0, 4], show=False) # graph of normal

p1.extend(p2)
p1.extend(p3)
p1.show()
```

### Question 3

#### 3a

```python
# curvature A
x = symbols('x')
h = x**2 + 3*x + 5
dh = diff(h, x)
d2h = diff(h, x, 2)
k = abs(d2h)/(1+dh**2)**Rational(3/2)
print(k.subs(x, 2))
```

#### 3b

```python
# curvature B
x = symbols('x')
h = tan(x)
dh = diff(h, x)
d2h = diff(h, x, 2)
k = abs(d2h)/(1+dh**2)**Rational(3/2)
print(k.subs(x, pi/3))
```

#### 3c

```python
# curvature C
x = symbols('x')
h = 7*x-1
dh = diff(h, x)
d2h = diff(h, x, 2)
k = abs(d2h)/(1+dh**2)**Rational(3/2)
print(k.subs(x, 5))
```

#### 3d

```python
# curvature D
x = symbols('x')
h = sqrt(25-x**2)
dh = diff(h, x)
d2h = diff(h, x, 2)
k = abs(d2h)/(1+dh**2)**Rational(3/2)
print(k.subs(x, 1))
```
