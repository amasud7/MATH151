```python
# Start your code here
x = symbols('x')

# print justification
f = sqrt(x**3 + x**2) * sin(pi/x) + 2
g = - sqrt(x**3 + x**2) + 2
h = sqrt(x**3 + x**2) + 2
print(f"{g} is the lower bound, \n {h} is the upper bound \n to use squeeze theorem to find \n limit of {f}")

p1  = plot(f, (x, -0.8, 0.8), show=False)
p2 = plot(g, (x, -0.8, 0.8), show=False)
p3 = plot(h, (x, -0.8, 0.8), show=False)

p1.extend(p2) # this plots p2 and p1 on the same graph
p1.extend(p3)
p1.show()
```

#### 1b

```python
# Start your code here
print("This is the limit of h: ", limit(h, x, 0))
print("This is the limit of g: ", limit(g, x, 0))
print("Because the limits of g and h are equal we are able to apply the squeeze theorem to find the limit of f.")
```

```python
# start code here
x = symbols('x')
f1 = -2
f2 = 2 * x - 5
f3 = (((x ** 2) - (2 * x) - 8) / (x - 4))
f4 = exp(((x - 4) * ln(7)))

#x = 1
print('The left sided limit as x approaches 1 is', limit(f1, x, 1))
print('The right sided limit as x approaches 1 is', limit(f2, x, 1))
print('Because these limits are not equal, the function is not continuous at x = 1')
print()

#x = 3
print('The left sided limit as x approaches 3 is', limit(f2, x, 3))
print('The right sided limit as x approaches 3 is', limit(f3, x, 3))
print('Because these limits are not equal, the function is not continuous at x = 3')
print()

#x = 5
print('The left sided limit as x approaches 5 is', limit(f3, x, 5))
print('The right sided limit as x approaches 5 is', limit(f4, x, 5))
print('Because these limits are equal and are also equal to the function value, the function is continuous at x = 5')
```
