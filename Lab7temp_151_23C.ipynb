### Question 1

#### 1a

```python
x = symbols('x')
f = (x**2 + 8*x + 5)/(x**2)

df = diff(f, x)

criticalVal = solve(df, x)

criticalVal.append(-1)
criticalVal.append(-3)
print(f"These are the critical values: {criticalVal}")

fVals = [f.subs(x, i) for i in criticalVal]
print(f"This is the absolute max: {max(fVals)}")
print(f"This is the absolute min: {min(fVals)}")
```

### Question 2

#### 2a

```python
x = symbols('x')
f = 4*x**3 - 3*x**2 + 2*x - 1
slope = (f.subs(x, 7) - f.subs(x, 2))/5
df = diff(f, x)
cVal = solve(df - slope, x)
print(f"This is the c value: {cVal[1]}")
```

#### 2b

```python
matplotlib notebook
```

```python
p1 = plot(f, (x,2,7), show=False)
yp = f.subs(x, cVal[1])
tanL = yp + slope*(x-cVal[1])
p2 = plot(tanL, (x,2,7), show = False)
secL = slope*x - 465
p3 = plot(secL, (x,2,7), show=False)
p1.extend(p2)
p1.extend(p3)
p1.show()
```

```python
x = symbols('x')
f1 = 8 - x**2
f2 = 5*exp((x-2)/2) - x
df1 = diff(f1, x)
cVal1 = solve(df1, x)
df2 = diff(f2, x)
cVal2 = solve(df2, x)
print(f"These are the critical values: {cVal1[0]}, {cVal2[0]}")
```

#### 3b

```python
critical_val1 = solve(df1, x)
critical_val1.append(-5)
critical_val1.append(5)
fVals1 = [f.subs(x, i) for i in critical_val1]


critical_val2 = solve(df2, x)
critical_val2.append(-5)
critical_val2.append(5)
fVals2 = [f.subs(x, i) for i in critical_val2]
values = fVals1 + fVals2
print(f"There is no absolute max in this range, but there is a absolute min at {min(values)}.")
```
