### Question 1

#### 1a

```python
t = symbols('t')

xt = exp(2*sin(t))
yt = exp(cos(t))

dx = diff(xt, t)
dy = diff(yt, t)

mx = dx.subs(t, pi/6)
my = dy.subs(t, pi/6)
slope = my/mx


x0 = xt.subs(t, pi/6)
y0 = yt.subs(t, pi/6)


# print correct format for tangent line
print(f"This is the tangent line for this function at pi/6:\ny-{y0}={slope}(x-{x0})")
```

#### 1b

```python
# horizontal when dy = 0
horizontal = solve(dy, t)
xh1 = xt.subs(t, horizontal[0])
yh1 = yt.subs(t, horizontal[0])
xh2 = xt.subs(t, horizontal[1])
yh2 = yt.subs(t, horizontal[1])
print(f"These are the points where there is a horizontal tangent:\n({xh1}, {yh1}) and ({xh2}, {yh2})")

#vertical when dx = 0
vert = solve(dx, t)
xv1 = xt.subs(t, vert[0])
yv1 = yt.subs(t, vert[0])
xv2 = xt.subs(t, vert[1])
yv2 = yt.subs(t, vert[1])
print(f"These are the points where there is a vertical tangent:\n({xv1}, {yv1}) and ({xv2}, {yv2})")

# make sure to print correct statement
```

### Question 2

#### 2a

```python
x, y = symbols('x y')
f = x**2 + 2*x + 3*y**2-8
dydx = idiff(f, y, x)
print(f"This is the derivative of the function:\n{dydx}")
```

#### 2b

```python
num, den = fraction(dydx)

horiz = solve([num, f], [x, y])
print(f"These are the points where there is a horizontal tangent: {horiz}")
```

#### 2c

```python
vert = solve([den, f], [x, y])
print(f"These are the points where there is a vertical tangent: {vert}")
```
