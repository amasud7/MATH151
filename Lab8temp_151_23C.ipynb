### Question 1

#### 1a

```python
x = symbols('x')
f = 1/40*(x**6 + 2 *x**5 - 16*x**4 - 20*x**3 + 64*x**2 - 36*x + 72)
df = diff(f, x)
crit = solve(df, x)
print(crit)

for val in crit:
    if df.subs(x, val-0.1) < 0:  
        print(f'The function at crit point {val} is {df.subs(x, val-0.1)} and is increasing')
    else:
        print(f'The function at crit point {val} is {df.subs(x, val-0.1)} and is decreasing')

print(f"The function at crit point {crit[4]} is {df.subs(x, crit[4]+0.1)} and is increasing")
plot(df, (x,-5,5), ylim = [-5,5])
```
