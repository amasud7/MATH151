```python
#start code here
t = symbols('t')

x1 = sin(2*t)
y1 = cos(t)

x2 = 3*sin(2*t)
y2 = cos(t)

x3 = sin(3*t)
y3 = 2*cos(t)

p = plot_parametric((x1, y1), (x2, y2), (x3, y3), (t, 0, 2*pi), show=False)
p[0].line_color = 'b'
p[1].line_color = 'g'
p[2].line_color = 'r'
p.show()
```
